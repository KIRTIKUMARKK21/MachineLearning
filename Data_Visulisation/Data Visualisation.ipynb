```python
x=np.linspace(0,10,1000)
y1=np.sin(x)
y2=np.cos(x)
y3=y2+y1
plt.plot(x,y1,'r')
plt.plot(x,y2,'b')
plt.plot(x,y1+y2,'black')
plt.xlim(-5,15)
plt.ylim(-3,3)
```

```python
plt.plot(x,y1,'-.r',label='sin(x)')
plt.plot(x,y2,'-g',label='cos(x)')
plt.plot(x,y3,':b',label='sin(x)+cos(x)')
plt.axis('tight')
plt.title("Sin and Cos Plots")
plt.xlabel("X")
plt.ylabel('sin(x),cos(x),sin(x)+cos(x)')
plt.legend()
```

```python
#contour plot
```

```python
x=np.linspace(0,10,50)
y=np.linspace(0,10,40)
X,Y=np.meshgrid(x,y)
Z=10*np.sin(X)+10*np.cos(Y)
plt.contour(X,Y,Z,30,cmap='RdGy')
plt.colorbar()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Conutour Plot")
```

```python
x=np.linspace(0,10,50)
y=np.linspace(0,10,40)
X,Y=np.meshgrid(x,y)
Z=10*np.sin(X)+2*Y
plt.contourf(X,Y,Z,30,cmap='RdGy')
plt.colorbar()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Conutour Plot")
```

```python
#histogram
```

```python
X=50*np.random.randn(5000)
Y=200*np.random.randn(5000)
Z=100*np.random.randn(5000)+500
plt.hist(X,density=True,alpha=0.5)
plt.hist(Y,density=True,alpha=0.5)
plt.hist(Z,density=True,alpha=0.5)
```

```python
import pandas as pd
```

```python
data=pd.DataFrame({'X':X,'Y':Y,'Z':Z})
```

```python
for col in data.columns:
    plt.hist(data[col],density=True,alpha=0.5)
    sns.kdeplot(data[col],shade=True)
```

```python
for col in data.columns:
    sns.distplot(data[col])
```

```python
for col in data.columns:
    sns.distplot(data[col])
```
